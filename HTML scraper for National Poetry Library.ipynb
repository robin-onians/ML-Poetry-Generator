```python
import requests

resp = requests.get("https://www.nationalpoetrylibrary.org.uk/online-poetry/poems/man-without-fear-0")
html_str = resp.text
```

```python
html_str
```

```python
from bs4 import BeautifulSoup

document = BeautifulSoup(html_str, "html.parser")
type(document)
```

```python
checkup_tags = document.find('div', attrs={'class': 'view-content'})
#[tag for tag in checkup_tags]

p = checkup_tags.find_all('p')
print(p)
type(p)

```

```python
import re
regex = re.compile("(?i)<[^>]*>")
#First parameter is the replacement, second parameter is the input string

x = []
for p_tag in p:
    x.append(p_tag.prettify())
    

y = []
for count in range(len(x)):
    t = x[count].split('<br/>')
    for line in t:
        #remove HTML tags
        regex = re.compile("(?i)<[^>]*>")
        i = regex.sub('', line)
        #remove \n occurences
        regex = re.compile("\n")
        j = regex.sub('', i)
        #strip punctuation
        j = j.strip(' ();:.,"!?  ')
        if j != "":
            y.append(j)

y
```
